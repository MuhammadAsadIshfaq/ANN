```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
lr = 0.5
nx = 2
n_H = 2
n_z = 2

wji = np.array([[0.15,0.20,0.35],[0.25,0.30,0.35]])

wkj = np.array([[0.40,0.45,0.60],[0.50,0.55,0.60]])

x = np.array([0.05,0.10])

z=np.array([0.01,0.99])

net_y1 = 0
net_y2 = 0
net_z1 = 0
net_z2 = 0

for i in range(nx):
    net_y1 += x[i]*wji[0][i]
    if i==0:
        net_y1 += wji[i][2]
    net_y2 += x[i]*wji[1][i]
    if i==1:
        net_y2 += wji[i][2]
        
y1 = sigmoid(net_y1)
y2 = sigmoid(net_y2)

for i in range(n_H):
    net_z1 += y1*wkj[0][i]
    if i==0:
        net_z1 += wkj[i][2]
    net_z2 += y2*wkj[1][i]
    if i==1:
        net_z2 += wkj[i][2]
        
z1 = sigmoid(net_z1)
z2 = sigmoid(net_z2)

E = 0.5*((z1-z[0])**2 + (z2-z[1])**2)

print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)
```

```python
# Sensitivity of Z1
S_z1 = (z[0]-z1)*(z1*(1-z1))

#delta wz1y1
d_wz1y1 = lr*S_z1*y1
#Update weight
wz1y1 = wkj[0][0] + d_wz1y1
#delta wz1y2
d_wz1y2 = lr*S_z1*y2
#Update weight
wz1y2 = wkj[0][1] + d_wz1y2
#delta bias
d_bz1 = lr*S_z1*1
# Update bias
bz1 = wkj[0][2] + d_bz1
```

```python
#Sensitivity of Z2
S_z2 = (z[1]-z2)*z2*(1-z2)

#delta wz2y1
d_wz2y1 = lr*S_z2*y1
#update weights
wz2y1 = wkj[1][0] + d_wz2y1
#delta wz2y2
d_wz2y2 = lr*S_z2*y2
#Update
wz2y2 = wkj[1][1] + d_wz2y2
# delta bias
d_bz2 = lr*S_z2*1
#Update
bz2 = wkj[1][2] + d_bz2
```

```python
# Sensitivity of y1
S_y1 = y1*(1-y1)*(S_z1*wkj[0][0] + S_z2*wkj[1][0])

#delta wy1x1
d_wy1x1 = lr * S_y1 * x[0]
#delta wy1x2
d_wy1x2 = lr * S_y1 * x[1]
#delta by1
d_by1 = lr * S_y1
#update wy1x1
wy1x1 = wji[0][0] - d_wy1x1
#update wy1x2
wy1x2 = wji[0][1] - d_wy1x2
#update by1
by1 = wji[0][2]-d_by1
```

```python
# Sensitivity of y2
S_y2 = y2*(1-y2)*(S_z1*wkj[0][1] + S_z2*wkj[1][1])

#delta wy2x1
d_wy2x1 = lr * S_y2 * x[0]
#delta wy2x2
d_wy2x2 = lr * S_y2 * x[1]
#delta by2
d_by2 = lr * S_y2
#update wy1x1
wy2x1 = wji[1][0] - d_wy2x1
#update wy1x2
wy2x2 = wji[1][1] - d_wy2x2
#update by1
by2 = wji[1][2] -d_by2
```

```python
#again feed foward with updated weights and bias
net_y1 = x[0] * wy1x1 + x[1]*wy1x2 + by1
y1 = 1/(1+np.exp(-net_y1))
net_y2 = x[0] * wy2x1 + x[1]*wy2x2 + by2
y2 = 1/(1+np.exp(-net_y2))

net_z1 = y1*wz1y1 + y2*wz1y2 +bz1
z1 = 1/(1+np.exp(-net_z1))
net_z2 = y1*wz2y1 + y2*wz2y2 +bz2
z2 = 1/(1+np.exp(-net_z2))

E = 0.5*((z1-z[0])**2 + (z2-z[1])**2)


print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

lr = 0.75
nx = 3
n_H = 2
n_z = 2

wji = np.array([[0.2, 0.3, -0.1], [0.1, -0.1, 0.2]])

wkj = np.array([[0.1, 0.5], [0.3, 0.2]])

x = np.array([1, 0.4, 0.7])

z = np.array([0.65, 0.65])

net_y1 = 0
net_y2 = 0
net_z1 = 0
net_z2 = 0

for i in range(nx):
    net_y1 += x[i] * wji[0][i]
    net_y2 += x[i] * wji[1][i]

y1 = sigmoid(net_y1)
y2 = sigmoid(net_y2)

for i in range(n_H):
    net_z1 += y1 * wkj[0][i]
    net_z2 += y2 * wkj[1][i]

z1 = sigmoid(net_z1)
z2 = sigmoid(net_z2)

E = 0.5 * ((z1 - z[0]) ** 2 + (z2 - z[1]) ** 2)

print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)

# Sensitivity of Z1
S_z1 = (z[0] - z1) * (z1 * (1 - z1))

# delta wz1y1
d_wz1y1 = lr * S_z1 * y1
# Update weight
wz1y1 = wkj[0][0] + d_wz1y1
# delta wz1y2
d_wz1y2 = lr * S_z1 * y2
# Update weight
wz1y2 = wkj[0][1] + d_wz1y2

# Sensitivity of Z2
S_z2 = (z[1] - z2) * z2 * (1 - z2)

# delta wz2y1
d_wz2y1 = lr * S_z2 * y1
# update weights
wz2y1 = wkj[1][0] + d_wz2y1
# delta wz2y2
d_wz2y2 = lr * S_z2 * y2
# Update
wz2y2 = wkj[1][1] + d_wz2y2

# Sensitivity of y1
S_y1 = y1 * (1 - y1) * (S_z1 * wkj[0][0] + S_z2 * wkj[1][0])

# delta wy1x1
d_wy1x1 = lr * S_y1 * x[0]
# delta wy1x2
d_wy1x2 = lr * S_y1 * x[1]
# delta wy1x3
d_wy1x3 = lr * S_y1 * x[2]

# Sensitivity of y2
S_y2 = y2 * (1 - y2) * (S_z1 * wkj[0][1] + S_z2 * wkj[1][1])

# delta wy2x1
d_wy2x1 = lr * S_y2 * x[0]
# delta wy2x2
d_wy2x2 = lr * S_y2 * x[1]
# delta wy2x3
d_wy2x3 = lr * S_y2 * x[2]

# Update weights and biases
wkj[0][0] = wz1y1
wkj[0][1] = wz1y2
wkj[1][0] = wz2y1
wkj[1][1] = wz2y2

# Update weights for y1
wji[0][0] += d_wy1x1
wji[0][1] += d_wy1x2
wji[0][2] += d_wy1x3

# Update weights for y2
wji[1][0] += d_wy2x1
wji[1][1] += d_wy2x2
wji[1][2] += d_wy2x3

print("Updated weights:")
print("wji:", wji)
print("wkj:", wkj)
```

```python
# Forward pass with updated weights and biases
net_y1 = np.dot(x, wji[0, :2]) + wji[0, 2]
y1 = sigmoid(net_y1)
net_y2 = np.dot(x, wji[1, :2]) + wji[1, 2]
y2 = sigmoid(net_y2)

net_z1 = np.dot([y1, y2], wkj[:, :1]) + wkj[0, 2]
z1 = sigmoid(net_z1)
net_z2 = np.dot([y1, y2], wkj[:, 1:]) + wkj[1, 2]
z2 = sigmoid(net_z2)

# Calculate new error
E = 0.5 * ((z1 - z[0]) ** 2 + (z2 - z[1]) ** 2)

print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)
```
