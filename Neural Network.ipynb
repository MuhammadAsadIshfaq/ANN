```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
lr = 0.5
nx = 2
n_H = 2
n_z = 2

wji = np.array([[0.15,0.20,0.35],[0.25,0.30,0.35]])

wkj = np.array([[0.40,0.45,0.60],[0.50,0.55,0.60]])

x = np.array([0.05,0.10])

z=np.array([0.01,0.99])

net_y1 = 0
net_y2 = 0
net_z1 = 0
net_z2 = 0

for i in range(nx):
    net_y1 += x[i]*wji[0][i]
    if i==0:
        net_y1 += wji[i][2]
    net_y2 += x[i]*wji[1][i]
    if i==1:
        net_y2 += wji[i][2]
        
y1 = sigmoid(net_y1)
y2 = sigmoid(net_y2)

for i in range(n_H):
    net_z1 += y1*wkj[0][i]
    if i==0:
        net_z1 += wkj[i][2]
    net_z2 += y2*wkj[1][i]
    if i==1:
        net_z2 += wkj[i][2]
        
z1 = sigmoid(net_z1)
z2 = sigmoid(net_z2)

E = 0.5*((z1-z[0])**2 + (z2-z[1])**2)

print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)
```

```python
# Sensitivity of Z1
S_z1 = (z[0]-z1)*(z1*(1-z1))

#delta wz1y1
d_wz1y1 = lr*S_z1*y1
#Update weight
wz1y1 = 0.40 + d_wz1y1
#delta wz1y2
d_wz1y2 = lr*S_z1*y2
#Update weight
wz1y2 = 0.45 + d_wz1y2
#delta bias
d_bz1 = lr*S_z1*1
# Update bias
bz1 = 0.60 + d_bz1
```

```python
#Sensitivity of Z2
S_z2 = (z[1]-z2)*z2*(1-z2)

#delta wz2y1
d_wz2y1 = lr*S_z2*y1
#update weights
wz2y1 = 0.50 + d_wz2y1
#delta wz2y2
d_wz2y2 = lr*S_z2*y2
#Update
wz2y2 = 0.55 + d_wz2y2
# delta bias
d_bz2 = lr*S_z2*1
#Update
bz2 = 0.60 + d_bz2
```

```python
# Sensitivity of y1
S_y1 = y1*(1-y1)*(S_z1*0.40 + S_z2*0.50)

#delta wy1x1
d_wy1x1 = 0.5 * S_y1 * 0.05
#delta wy1x2
d_wy1x2 = 0.5 * S_y1 * 0.10
#delta by1
d_by1 = 0.5 * S_y1
#update wy1x1
wy1x1 = 0.15 - d_wy1x1
#update wy1x2
wy1x2 = 0.20 - d_wy1x2
#update by1
by1 = 0.35-d_by1
```

```python
# Sensitivity of y2
S_y2 = y2*(1-y2)*(S_z1*0.45 + S_z2*0.55)

#delta wy2x1
d_wy2x1 = 0.5 * S_y2 * 0.05
#delta wy2x2
d_wy2x2 = 0.5 * S_y2 * 0.10
#delta by2
d_by2 = 0.5 * S_y2
#update wy1x1
wy2x1 = 0.25 - d_wy2x1
#update wy1x2
wy2x2 = 0.30 - d_wy2x2
#update by1
by2 = 0.35-d_by2
```

```python
#again feed foward with updated weights and bias
net_y1 = x[0] * wy1x1 + x[1]*wy1x2 + by1
y1 = 1/(1+np.exp(-net_y1))
net_y2 = x[0] * wy2x1 + x[1]*wy2x2 + by2
y2 = 1/(1+np.exp(-net_y2))

net_z1 = y1*wz1y1 + y2*wz1y2 +bz1
z1 = 1/(1+np.exp(-net_z1))
net_z2 = y1*wz2y1 + y2*wz2y2 +bz2
z2 = 1/(1+np.exp(-net_z2))

E = 0.5*((z1-z[0])**2 + (z2-z[1])**2)


print("net_y1:", net_y1)
print("net_y2:", net_y2)
print("y1:", y1)
print("y2:", y2)
print("net_z1:", net_z1)
print("net_z2:", net_z2)
print("z1:", z1)
print("z2:", z2)
print("Error:", E)
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def feedforward(inputs, weights, biases):
    net_inputs = np.dot(weights, inputs) + biases
    return sigmoid(net_inputs)

def backpropagation(inputs, targets, outputs, weights, lr):
    error = targets - outputs
    delta = error * outputs * (1 - outputs)
    delta_weights = lr * np.outer(delta, inputs)
    new_weights = weights + delta_weights
    return new_weights

def main():
    nx = int(input("Enter the number of input neurons: "))
    n_H = int(input("Enter the number of hidden neurons: "))
    n_z = int(input("Enter the number of output neurons: "))
    lr = float(input("Enter the learning rate: "))

    # Initialize weights and biases
    wji = np.random.rand(n_H, nx)
    wkj = np.random.rand(n_z, n_H)
    biases_j = np.random.rand(n_H)
    biases_k = np.random.rand(n_z)

    # Input values
    x = np.array([float(input(f"Enter value for x[{i + 1}]: ")) for i in range(nx)])
    z_target = np.array([float(input(f"Enter target value for z[{i + 1}]: ")) for i in range(n_z)])

    # Feedforward
    y = feedforward(x, wji, biases_j)
    z = feedforward(y, wkj, biases_k)

    # Calculate initial error
    error = 0.5 * np.sum((z - z_target) ** 2)
    print("Initial Error:", error)

    # Backpropagation
    delta_k = (z - z_target) * z * (1 - z)
    delta_j = y * (1 - y) * np.dot(wkj.T, delta_k)

    # Update weights and biases
    wkj = backpropagation(y, z_target, z, wkj, lr)
    wji = backpropagation(x, delta_j, y, wji, lr)
    biases_k -= lr * delta_k
    biases_j -= lr * delta_j

    # Feedforward again with updated weights and biases
    y = feedforward(x, wji, biases_j)
    z = feedforward(y, wkj, biases_k)

    # Calculate updated error
    error = 0.5 * np.sum((z - z_target) ** 2)
    print("Updated Error:", error)

if __name__ == "__main__":
    main()
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def feedforward(inputs, weights, biases):
    net_inputs = np.dot(weights, inputs) + biases
    return sigmoid(net_inputs)

def backpropagation(inputs, targets, outputs, weights, lr):
    error = targets - outputs
    delta = error * outputs * (1 - outputs)
    delta_weights = lr * np.outer(delta, inputs)
    new_weights = weights + delta_weights
    return new_weights

def train_multiple_times(n_iter):
    nx = int(input("Enter the number of input neurons: "))
    n_H = int(input("Enter the number of hidden neurons: "))
    n_z = int(input("Enter the number of output neurons: "))
    lr = float(input("Enter the learning rate: "))

    # Initialize weights and biases
    wji = np.zeros((n_H, nx))
    wkj = np.zeros((n_z, n_H))
    biases_j = np.zeros(n_H)
    biases_k = np.zeros(n_z)

    # Input values for weights wji
    print("Enter weights for wji:")
    for i in range(n_H):
        for j in range(nx):
            wji[i][j] = float(input(f"wji[{i + 1}][{j + 1}]: "))

    # Input values for weights wkj
    print("Enter weights for wkj:")
    for i in range(n_z):
        for j in range(n_H):
            wkj[i][j] = float(input(f"wkj[{i + 1}][{j + 1}]: "))

    # Input values for biases biases_j
    print("Enter biases for hidden layer:")
    for i in range(n_H):
        biases_j[i] = float(input(f"biases_j[{i + 1}]: "))

    # Input values for biases biases_k
    print("Enter biases for output layer:")
    for i in range(n_z):
        biases_k[i] = float(input(f"biases_k[{i + 1}]: "))

    # Input values
    x = np.array([float(input(f"Enter value for x[{i + 1}]: ")) for i in range(nx)])
    z_target = np.array([float(input(f"Enter target value for z[{i + 1}]: ")) for i in range(n_z)])

    for _ in range(n_iter):
        # Feedforward
        y = feedforward(x, wji, biases_j)
        z = feedforward(y, wkj, biases_k)
        print("The value of z is :",z)

        # Backpropagation
        delta_k = (z - z_target) * z * (1 - z)
        delta_j = y * (1 - y) * np.dot(wkj.T, delta_k)

        # Update weights and biases
        wkj = backpropagation(y, z_target, z, wkj, lr)
        wji = backpropagation(x, delta_j, y, wji, lr)
        biases_k -= lr * delta_k
        biases_j -= lr * delta_j

        # Feedforward again with updated weights and biases
        y = feedforward(x, wji, biases_j)
        z = feedforward(y, wkj, biases_k)
        print("The value of z updated is :",z)


        # Calculate error
        error = 0.5 * np.sum((z - z_target) ** 2)
        print("Error:", error)

if __name__ == "__main__":
    n_iterations = int(input("Enter the number of iterations: "))
    train_multiple_times(n_iterations)
```
